```python
from itertools import accumulate
import cruise
import torch
import torch.nn.functional as F
from megatron.core import parallel_state as mpu
from megatron.core.tensor_parallel.mappings import (
    reduce_scatter_to_sequence_parallel_region,
)
from megatron.model.module import MegatronModule
from mariana.utils.env_config import get_env_config
from mariana.utils.others import mrank_zero_info
# activation norm
report_activation_per_n_layer = get_env_config().log_per_n_layer_activation_norm
report_accurate_activation = get_env_config().report_accurate_activation_norm
check_moe_tensor_diff = get_env_config().check_moe_tensor_diff
def count_expert(num_experts, level=4):
    if num_experts == 0:
        return 0
    group_sizes = [num_experts // level * 2**i for i in range(level)]
    count = 0
    for _, group_size in enumerate(group_sizes):
        count += group_size
    return count
def dispatch(inp, top_k_gates, top_k_indices, num_experts):
    """Dispatch input to experts"""
    token_idx = (
        torch.arange(0, inp.shape[0], device=inp.device, dtype=top_k_indices.dtype)
        .unsqueeze(1)
        .repeat(1, top_k_indices.shape[-1])
        .flatten()
    )
    top_k_gates_flatten = top_k_gates.reshape(-1, 1)
    top_k_indices_flatten = top_k_indices.flatten()
    _, index_sorted_experts = top_k_indices_flatten.sort()
    pack_index = token_idx[index_sorted_experts]
    # 获取每个expert处理多少个token，长度为expert个数
    with torch.no_grad():
        expert_ids, _part_sizes = top_k_indices_flatten.unique(return_counts=True)
        part_sizes = _part_sizes.tolist()
    pack_gates = top_k_gates_flatten[index_sorted_experts, :]
    # 按照expert把token特征摞一块
    inp_exp = inp[pack_index].squeeze(1)
    return (
        torch.split(inp_exp, part_sizes, dim=0),
        pack_gates,
        pack_index,
        expert_ids,
        part_sizes,
    )
def combine(expert_out, x, pack_gates, pack_index, multiply_by_gates=True):
    """Combine expert activations into one"""
    stitched = torch.cat(expert_out, 0)
    if multiply_by_gates:
        stitched = stitched.mul(pack_gates)
    zeros = torch.zeros(x.size(0), expert_out[-1].size(1), device=stitched.device, dtype=stitched.dtype)
    # combine samples that have been processed by the same k experts
    combined = zeros.index_add(0, pack_index, stitched)
    return combined
def expert_to_gates(nonzero_gates, part_sizes):
    """Gate values corresponding to the examples in the per-expert `Tensor`s.
    Returns:
        a list of `num_experts` one-dimensional `Tensor`s with type `torch.float32`
            and shapes `[expert_batch_size_i]`
    """
    # split nonzero gates for each expert
    return torch.split(nonzero_gates, part_sizes, dim=0)
class MLP(torch.nn.Module):
    def __init__(self, input_size, output_size, hidden_size):
        super().__init__()
        self.fc1 = torch.nn.Linear(input_size, hidden_size)
        self.fc2 = torch.nn.Linear(hidden_size, output_size)
        self.gelu = F.gelu
    def forward(self, x):
        out = self.fc1(x)
        out = self.gelu(out)
        out = self.fc2(out)
        return out
class AuxLossBackwardHook(torch.autograd.Function):
    @staticmethod
    def forward(ctx, output, aux_loss, scale):
        # Preserve the aux_loss by storing it in the context to avoid garbage collection.
        ctx.save_for_backward(aux_loss)
        ctx.scale = scale
        return output
    @staticmethod
    def backward(ctx, grad_output):
        # Scale the auxiliary loss like the main loss.
        (aux_loss,) = ctx.saved_tensors
        scaled_aux_loss_grad = torch.ones_like(aux_loss) * ctx.scale
        return grad_output, scaled_aux_loss_grad, None
class NaiveMoE(torch.nn.Module):
    """Call a Sparsely gated mixture of experts layer with 1-layer Feed-Forward networks as experts.
    Args:
      input_size: integer - size of the input
      output_size: integer - size of the input
      num_experts: an integer - number of experts
      hidden_size: an integer - hidden size of the experts
      noisy_gating: a boolean
      k: an integer - how many experts to use for each batch element
    """
    def __init__(
        self,
        layer_number,
        input_size,
        output_size,
        num_experts,
        hidden_size,
        noisy_gating=True,
        k=2,
        level=4,
        alpha=0.999,
        ce_alpha=0.0,
        aux_coef=0.001,
        over_compute=1.0,
    ):
        super().__init__()
        self.experts = torch.nn.ModuleList()
        # 2, 4, 8, 16, ....
        self.group_sizes = [num_experts // level * 2**i for i in range(level)]
        self.group_index_offsets = list(accumulate([0] + self.group_sizes))[:-1]
        self.expert_caps = []
        for i, group_size in enumerate(self.group_sizes):
            self.experts.extend([MLP(input_size, output_size, hidden_size // (2**i)) for _ in range(group_size)])
            self.expert_caps.extend([1.0 / (2**i) for _ in range(group_size)])
        self.register_buffer("cal_weights", torch.tensor(self.expert_caps))
        self.layer_number = layer_number
        self.num_experts = len(self.experts)
        total_num_experts = count_expert(num_experts, level)
        assert self.num_experts == total_num_experts, (
            self.num_experts,
            total_num_experts,
        )
        self.output_size = output_size
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.k = k
        self.alpha = alpha
        self.level = level
        self.group_size = group_size
        self.aux_coef = aux_coef
        self.over_compute = over_compute
        self.ce_alpha = ce_alpha
        self.w_gate = torch.nn.Parameter(torch.zeros(input_size, self.num_experts), requires_grad=True)
        self.register_buffer("w_gate_ema", torch.zeros(input_size, self.num_experts))
        self.register_buffer("ce_ema", torch.zeros(self.num_experts))
        moe_info = f"num_experts={self.num_experts} top_k={self.k} expert_caps={self.expert_caps} "
        moe_info += f"group_sizes={self.group_sizes} group_index_offsets={self.group_index_offsets}"
        if cruise.utilities.DIST_ENV.local_rank == 0:
            print(moe_info)
        self.softmax = torch.nn.Softmax(1)
        assert self.k <= self.num_experts
    def moe_index_hist(self, top_k_indices, aux_loss, batch_tokens):
        trainer = cruise.last_cli().trainer
        with torch.no_grad():
            metric_name = f"layer_{self.layer_number}_aux_loss"
            trainer.log(metric_name, aux_loss)
            cap = 0
            for idx in range(self.num_experts):
                ones = torch.ones_like(top_k_indices, dtype=torch.float32)
                zeros = torch.zeros_like(top_k_indices, dtype=torch.float32)
                per_expert_index = torch.where(torch.eq(top_k_indices, idx), ones, zeros)
                ratio = torch.mean(per_expert_index.float().sum(dim=-1))
                metric_name = f"moe_layer_{self.layer_number}_e_{idx}_cap"
                cap_per_expert = ratio * self.expert_caps[idx]
                cap += cap_per_expert
                trainer.log(metric_name, cap_per_expert)
            metric_name = f"moe_same_index_layer_{self.layer_number}_cap"
            trainer.log(metric_name, cap)
            # The number of tokens equivalent to the Dense model under MoE
            moe_equiv_tokens = cap * batch_tokens
        return moe_equiv_tokens
    def cv_squared(self, x):
        """The squared coefficient of variation of a sample.
        Useful as a loss to encourage a positive distribution to be more uniform.
        Epsilons added for numerical stability.
        Returns 0 for an empty Tensor.
        Args:
          x: a `Tensor`.
        Returns:
          a `Scalar`.
        """
        eps = 1e-10
        # if only num_experts = 1
        if x.shape[0] == 1:
            return torch.tensor([0], device=x.device, dtype=x.dtype)
        return x.float().var() / (x.float().mean() ** 2 + eps)
    def _gates_to_load(self, gates):
        """Compute the true load per expert, given the gates.
        The load is the number of examples for which the corresponding gate is >0.
        Args:
          gates: a `Tensor` of shape [batch_size, n]
        Returns:
          a float32 `Tensor` of shape [n]
        """
        return (gates > 0).sum(0)
    def group_top_1_gating(self, x):
        logits = self.gate_net(x)
        logits_group = torch.split(logits, self.group_sizes, dim=-1)
        top_k_logits_group = []
        top_k_indices_group = []
        for offset, _logits in zip(self.group_index_offsets, logits_group):
            top_k_logits, top_k_indices = _logits.max(dim=-1)
            top_k_logits_group.append(top_k_logits)
            top_k_indices_group.append(top_k_indices + offset)
        top_k_indices = torch.stack(top_k_indices_group, dim=-1)
        top_k_logits = torch.stack(top_k_logits_group, dim=-1)
        top_k_gates = self.softmax(top_k_logits)
        with torch.no_grad():
            self.w_gate_ema.copy_(self.alpha * self.w_gate_ema + (1 - self.alpha) * self.w_gate)
        self.moe_index_hist(top_k_indices)
        return top_k_gates
    def top_k_gating(self, x):
        logits = x @ self.w_gate
        top_k_logits, top_k_indices = logits.topk(self.k, dim=1)
        top_k_gates = self.softmax(top_k_logits)
        cap = self.moe_index_hist(top_k_indices)
        return top_k_gates, top_k_indices, cap
    def ema_top_k_gating(self, x):
        w_gate_running = 0.5 * (self.w_gate_ema.bfloat16() + self.w_gate)
        logits = x @ w_gate_running
        top_k_logits, top_k_indices = logits.topk(self.k, dim=1)
        top_k_gates = self.softmax(top_k_logits)
        mask = F.one_hot(top_k_indices, num_classes=self.num_experts).sum(dim=1)
        # Compute l_aux
        nme = self.softmax(-logits).mean(dim=0)
        ce = mask.float().mean(dim=0)
        ce = ce * self.cal_weights
        with torch.no_grad():
            self.w_gate_ema = self.alpha * self.w_gate_ema.float() + (1 - self.alpha) * self.w_gate.float()
            self.ce_ema = self.ce_alpha * self.ce_ema.float() + (1 - self.ce_alpha) * ce.float()
        hot_num, hot_exp = self.ce_ema.max(), self.ce_ema.argmax()
        cold_num, _ = self.ce_ema.min(), self.ce_ema.argmin()
        l_aux = -F.relu(hot_num - cold_num * self.over_compute) * nme[hot_exp].log()
        with torch.no_grad():
            moe_experts_cap = ce.clone().float()
            moe_cap = moe_experts_cap.sum()
        return top_k_gates, top_k_indices, l_aux, [moe_cap, moe_experts_cap]
    def forward(self, x, loss_coef=1e-2):
        top_k_gates, top_k_indices, l_aux, moe_cap = self.ema_top_k_gating(x)
        # 把tokens 按照 expert分组
        expert_inputs, pack_gates, pack_index, expert_ids, part_sizes = dispatch(
            x, top_k_gates, top_k_indices, self.num_experts
        )
        # 分批过expert, 注意part_sizes[i]=0要跳过
        expert_outputs = [self.experts[expert_id](expert_inputs[i]) for i, expert_id in enumerate(expert_ids)]
        # 把结果再组合回来
        y = combine(expert_outputs, x, pack_gates, pack_index)
        y = AuxLossBackwardHook.apply(y, l_aux, self.aux_coef)
        return y, l_aux, moe_cap
class ParallelMOE(MegatronModule):
    """MOE Layer.
    Sparsely gated mixture of experts layer with 1-layer Feed-Forward networks as experts.
    backend: default, janus
    Arguments
    ---------
    layer_number: current layer number
    moe_overlap_recomp_grad_comm: Bool. Default to False. Enable overlapping the allgather of grad_output
    model_config: configuration options for the model
    megatron_config: onfiguration options for Megatron training
    """
    def __init__(
        self,
        layer_number,
        model_config=None,
        megatron_config=None,
    ):
        super().__init__()
        self.model_config = model_config
        self.megatron_config = megatron_config
        self.layer_number = max(1, layer_number)
        # MOE
        topk = self.model_config.get('moe_topk', 0)
        self.moe_backend = self.model_config.get("moe_backend", "default")
        target_inner_dim = self.model_config.get_mlp_inner_dim(self.layer_number)
        if layer_number == 1:
            mrank_zero_info(f"Use model.network.n_inner={target_inner_dim} as moe FFN inner dim")
        if self.moe_backend == 'default':
            self.moe = NaiveMoE(
                self.layer_number,
                self.model_config.hidden_size,
                self.model_config.hidden_size,
                self.model_config.moe_num_expert,
                target_inner_dim,
                noisy_gating=False,
                k=topk,
            )
        elif self.moe_backend == 'janus':
            from janus.megatron import MegatronLegoMoE as JanusMoE
            moe_expert_type = self.model_config.get('moe_expert_type', 'exp-xelego')
            moe_gate_type = self.model_config.get('moe_gate_type', 'caplog-lego')
            moe_gate_metric_type = self.model_config.get('moe_gate_metric_type', 'lego')
            moe_expert_exp_level = self.model_config.get('moe_expert_exp_level', 4)
            moe_expert_exp_first_dim_factor = self.model_config.get('moe_expert_exp_first_dim_factor', 1.0)
            moe_expert_exp_first_num = self.model_config.get('moe_expert_exp_first_num', 2)
            moe_expert_eq_dim_factor = self.model_config.get('moe_expert_eq_dim_factor', 0.25)
            moe_expert_op_version = self.model_config.get('moe_expert_op_version', 'V1')
            moe_aux_loss_weight = self.model_config.get('moe_aux_loss_weight', 0.001)
            convert_gate_to_fp32 = self.model_config.get('convert_gate_to_fp32', False)
            gate_dropout = self.model_config.get('moe_gate_dropout', 0.0)
            enable_ema_update = self.model_config.get('moe_enable_ema_update', 1) > 0
            moe_use_balance = model_config.get('moe_use_balance', False)
            moe_expert_group_capacity = model_config.get('moe_expert_group_capacity', False)
            moe_expert_group_balance_loss_weight = model_config.get('moe_expert_group_balance_loss_weight', False)
            moe_expert_groups_in_ep_rank = model_config.get('moe_expert_groups_in_ep_rank', 1)
            moe_pr_scale_factor = model_config.get('moe_pr_scale_factor', 1.0)
            moe_pr_expert_type = model_config.get('moe_pr_expert_type', 'disabled')
            moe_swiglu_fc1_2_init_scale = model_config.get('moe_swiglu_fc1_2_init_scale', 1.0)
            hidden_decoding_layers = model_config.get('hidden_decoding_layers', [])
            if layer_number in hidden_decoding_layers:
                ep_size = mpu.get_expert_model_parallel_world_size()
                real_experts = max(ep_size, self.model_config.moe_num_expert)
            else:
                real_experts = self.model_config.moe_num_expert
            moe_routed_scaling_factor = model_config.get('moe_routed_scaling_factor', 1.0)
            moe_renorm = model_config.get('moe_renorm', True)
            self.moe = JanusMoE(
                self.layer_number,
                self.model_config.hidden_size,
                real_experts,
                target_inner_dim,
                topk,
                aux_loss_weight=moe_aux_loss_weight,
                expert_type=moe_expert_type,
                pr_expert_type=moe_pr_expert_type,
                gate_type=moe_gate_type,
                gate_metric_type=moe_gate_metric_type,
                expert_exp_level=moe_expert_exp_level,
                expert_exp_first_dim_factor=moe_expert_exp_first_dim_factor,
                expert_exp_first_num=moe_expert_exp_first_num,
                expert_eq_dim_factor=moe_expert_eq_dim_factor,
                overlap_recomp_grad_comm=self.model_config.moe_overlap_recomp_grad_comm,
                expert_op_version=moe_expert_op_version,
                convert_gate_to_fp32=convert_gate_to_fp32,
                gate_dropout=gate_dropout,
                enable_ema_update=enable_ema_update,
                quant_real_store=self.megatron_config.qlora_quant_real_store,
                quant_aware_L4Q=self.megatron_config.qlora_quant_aware_L4Q,
                quant_groupsize=self.megatron_config.qlora_quant_groupsize,
                quant_weight_dtype=self.megatron_config.qlora_quant_weight_dtype,
                quant_input_dtype=self.megatron_config.qlora_quant_input_dtype,
                quant_aware_lora=self.megatron_config.qlora_quant_aware_lora,
                use_balance=moe_use_balance,
                expert_group_capacity=moe_expert_group_capacity,
                expert_group_balance_loss_weight=moe_expert_group_balance_loss_weight,
                expert_groups_in_ep_rank=moe_expert_groups_in_ep_rank,
                pr_scale_factor=moe_pr_scale_factor,
                swiglu_fc1_2_init_scale=moe_swiglu_fc1_2_init_scale,
                moe_routed_scaling_factor=moe_routed_scaling_factor,
                moe_renorm=moe_renorm,
                device_limit_topk=self.model_config.moe_device_limit_topk,
                comm_groups=self.model_config.moe_comm_groups,
                device_top_expert=self.model_config.moe_device_top_expert,
                gradient_accumulation_fusion=self.megatron_config.gradient_accumulation_fusion,
            )
        self.recompute_granularity = self.megatron_config.recompute_granularity
        self.num_layers = self.model_config.n_layer
        self.use_non_sequential_block = self.megatron_config.use_non_sequential_block
        self._num_invocation = 0
        self.moe_enable_warmup = model_config.get('moe_enable_warmup', False)
    def _get_warmup_rate(self):
        if self.moe_enable_warmup:
            trainer = cruise.last_cli().trainer
            global_config = cruise.last_cli().hparams
            global_step = trainer.global_step
            total_steps = trainer.total_steps
            use_bsz_warmup = global_config['data']['bsz_warmup']
            bsz_warmup_rate = global_config['data']['bsz_warmup_rate']
            if use_bsz_warmup:
                step_line = int(total_steps * bsz_warmup_rate)
                if global_step >= step_line:
                    warmup_rate = 1.0
                else:
                    warmup_rate = global_step / step_line
            else:
                warmup_rate = 1.0
        else:
            warmup_rate = 1.0
        return warmup_rate
    def forward(self, hidden_states, ret_activation_norm=False, cu_seqlens=None):
        b, l, d = hidden_states.shape
        if self.moe_backend == 'default':
            output, l_aux, cap = self.moe(hidden_states.reshape([b * l, d]))
            output = output.reshape([b, l, d])
            moe_cap, moe_experts_cap = cap
            self.moe_experts_cap = moe_experts_cap.detach()
        elif self.moe_backend == 'janus':
            if check_moe_tensor_diff:
                from janus.megatron.hook import DumpTensorStateHook
                root_dir = '/opt/tiger/moe_init_data/'
                rank = get_env_config().rank
                pp_stage = get_env_config().moe_init_pp_stage
                self._num_invocation += 1
                if self._num_invocation >= 2:
                    hidden_states = DumpTensorStateHook.apply(
                        hidden_states,
                        root_dir,
                        pp_stage,
                        self._num_invocation,
                        self.layer_number,
                        rank,
                        'hidden_states',
                    )
            warmup_rate = self._get_warmup_rate()
            output, l_aux, cap = self.moe(hidden_states, cu_seqlens=cu_seqlens, warmup_rate=warmup_rate)
            if check_moe_tensor_diff and self._num_invocation >= 2:
                output = DumpTensorStateHook.apply(
                    output, root_dir, pp_stage, self._num_invocation, self.layer_number, rank, 'output'
```
